```python
import zipfile
import os
import pandas as pd
```

```python
df = pd.read_excel('../학습데이터 리스트.xlsx')
df = df[['C-Code']]

zip_root = r'C:\Users\daniel\Capstone\AI\166.약품식별 인공지능 개발을 위한 경구약제 이미지 데이터\01.데이터\1.Training\라벨링데이터\단일경구약제 5000종'
```

```python
for file in os.listdir(zip_root):
    if file.endswith('.zip'):
        zip_path = os.path.join(zip_root, file)
        print(f"Processing: {file}")
        try:
            with zipfile.ZipFile(zip_path, 'r') as z:
                all_files = z.namelist()
                zip_folders = {f.split('/')[0] for f in all_files if '/' in f} # 디렉토리 폴더 이름 추출

                for i, row in df.iterrows():
                    ccode = row['C-Code']
                    for zip_folder in zip_folders:
                        if ccode in zip_folder: # C-Code가 zip 폴더 이름과 같은지 확인인
                            file_count = sum(1 for f in all_files if f.endswith('json') and f.startswith(ccode)) # json 파일 개수 세기
                            df.at[i, '데이터 개수'] = file_count
                            df.at[i, '데이터 위치'] = os.path.splitext(file)[0]  # zip 이름에서 .zip 제거
        except zipfile.BadZipFile:
            print(f"Bad zip file: {file}")
```
